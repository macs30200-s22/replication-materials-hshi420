```python
import pandas as pd
```

```python
df = pd.read_csv('politifact.csv', index_col = 0)
```

```python
df.head()
```

```python
meter = ['mostly true', 'half true', 'true', 'mostly false', 'false', 
         'pants on fire', 'full flop', 'no flip', 'half flip', 'half-flip', 'full-flip', 'no-flip']
```

```python
rating = []
for row in df.iloc:
    article = row.content.lower()
    flag = 0
    for rate in meter:
        if rate in article:
            flag = 1
            rating.append(rate)
            break
    if flag == 0:
        rating.append('none')
```

```python
df['label'] = rating
```

```python
df = df[df['label'] != 'none']
```

```python
df
```

```python
revised_label = []
true = ['true', 'mostly true']
false = ['half true', 'mostly false', 'false', 'pants on fire']
for row in df.iloc:
    if row.label in true:
        revised_label.append('true')
    elif row.label in false:
        revised_label.append('false')
    else:
        revised_label.append('none')
len(revised_label)
```

```python
revised_label.count('none')
```

```python
df['label'] = revised_label
```

```python
df = df[df['label'] != 'none']
```

```python
df.label.value_counts()
```

```python
df.head()
```

```python
len(df)
```

```python
personality = pd.read_csv('personality.csv', index_col = 0)
```

```python
personality.head()
```

```python
merged = pd.merge(df, personality, left_on = 'speaker_info', right_on = 'link')
```

```python
len(merged)
```

```python
df = merged[['speaker_x', 'label', 'factchecker','party']]
```

```python
df = df[df['party'] != 'None']
```

```python
df['democrat'] = 0
df['republican'] = 0
```

```python
df.loc[df.party == 'Democrat', 'democrat'] = 1
df.loc[df.party == 'Republican', 'republican'] = 1
```

```python
df
```

```python
df['democrat_false'] = 0
df['republican_false'] = 0
```

```python
df.loc[(df.democrat == 1) &(df.label == 'false'), 'democrat_false'] = 1
df.loc[(df.republican == 1) &(df.label == 'false'), 'republican_false'] = 1
```

```python
r = df[['factchecker', 'democrat_false', 'republican_false']]
```

```python
r['sum'] = r['democrat_false'] + r['republican_false']
```

```python
r = r.groupby('factchecker').mean()
```

```python
r = r.loc[r['sum'] > 0]
```

```python
r['democrat_false'].corr(r['republican_false'])
```

```python
import matplotlib.pyplot as plt
import numpy as np
#plt.scatter(r['democrat_false'], r['republican_false'])

from scipy import stats

coef = np.polyfit(r.democrat_false, r.republican_false, 1)

poly1d_fn = np.poly1d(coef)

plt.plot(r['democrat_false'],r['republican_false'], 'ro', r['republican_false'], poly1d_fn(r['republican_false']))

plt.xlabel('proportion of democrat statement rated as false')
plt.ylabel('proportion of republican statement rated as false')
```

```python
r['left_leaning'] = 0
```

```python
r.loc[r.democrat_false < r.republican_false, 'left_leaning'] = 1
```

```python
r = r.reset_index()
```

```python
r
```

```python
r.left_leaning.value_counts()
```

```python
r.to_csv('left_leaning_factcheckers.csv')
```
